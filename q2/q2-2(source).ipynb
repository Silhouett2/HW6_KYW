```python
import csv 
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

f = open('random.csv', 'r',encoding='UTF8')
data = csv.reader(f)


def readdata(datalist) :
    for i in range(len(datalist)):
        datalist[i] = int(datalist[i])
    return datalist

for i in range(4):
    data1 = next(data)
    d1 = readdata(data1)
    plt.hist(d1,bins=6)
    plt.title("주사위 시뮬레이션")
    plt.xlabel("dice")
    plt.ylabel("times of dice")
    plt.show()



# d1 = readdata(data1)
# d2 = readdata(next(data))
# d3 = readdata(next(data))
# d4 = readdata(next(data))
# plt.hist(d1)

# mapd1 = map(int, d1)
# mapd2 = map(int, d2)
# mapd3 = map(int, d3)
# mapd4 = map(int, d4)


# plt.hist(int(d1))
# plt.show()
# plt.hist(int(d1))
# plt.show()
# plt.hist(int(d1))
# plt.show()


f.close()



def main():
    print(" ")
if __name__ == '__main__':
    main()
```
