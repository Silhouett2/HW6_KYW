```python
import matplotlib.pyplot as plt
import csv
plt.rc('font', family='Malgun Gothic')

# 이거 나중에 함수로 동적변수? 만들어서 해보자

j2022 = open('2022제주.csv', 'r', encoding='cp949')
d2022 = csv.reader(j2022, delimiter=',')
j2017 = open('2017제주.csv', 'r', encoding='cp949')
d2017 = csv.reader(j2017, delimiter=',')
j2012 = open('2012제주.csv', 'r', encoding='cp949')
d2012 = csv.reader(j2012, delimiter=',')
j2008 = open('2008제주.csv', 'r', encoding='cp949')
d2008 = csv.reader(j2008, delimiter=',')


header = next(d2022)
p2022 = next(d2022)

# 숫자사이에 , 제거 해야되는데


p2022[14] = p2022[14].replace(',','')
man2022 = float(p2022[14])
p2022[28] = p2022[28].replace(',','')
woman2022 = float(p2022[28])

label = ["man","woman"]
ydata = [man2022,woman2022]

plt.pie(ydata,labels=label,autopct='%.2f%%')
plt.title("2022제주도 성비")
plt.legend()
plt.show()


header = next(d2017)
p2017 = next(d2017)

p2017[14] = p2017[14].replace(',','')
man2017 = float(p2017[14])
p2017[28] = p2017[28].replace(',','')
woman2017 = float(p2017[28])

label = ["man","woman"]
ydata = [man2017,woman2017]

plt.pie(ydata,labels=label,autopct='%.2f%%')
plt.title("2017제주도 성비")
plt.legend()
plt.show()


header = next(d2012)
p2012 = next(d2012)

p2012[14] = p2012[14].replace(',','')
man2012 = float(p2012[14])
p2012[28] = p2012[28].replace(',','')
woman2012 = float(p2012[28])

label = ["man","woman"]
ydata = [man2012,woman2012]

plt.pie(ydata,labels=label,autopct='%.2f%%')
plt.title("2012제주도 성비")
plt.legend()
plt.show()





j2022.close()
j2017.close()
j2012.close()
j2008.close()

def main():
    print(" ")
if __name__ == '__main__':
    main()
```
