```python
import matplotlib.pyplot as plt
import csv
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

koreafile = open('korea.csv', 'r', encoding='cp949')
koreadata = csv.reader(koreafile, delimiter=',')
seoulfile = open('seoul.csv', 'r', encoding='cp949')
seouldata = csv.reader(seoulfile, delimiter=',')
daejeonfile = open('daejeon.csv', 'r', encoding='cp949')
daejeondata = csv.reader(daejeonfile, delimiter=',')
busanfile = open('busan.csv', 'r', encoding='cp949')
busandata = csv.reader(busanfile, delimiter=',')
jejufile = open('jeju.csv', 'r', encoding='cp949')
jejudata = csv.reader(jejufile, delimiter=',')



place = []
global n
n = 0

def pcodechange (row1) :
    
    for i in range(len(row1)):
        if row1 == '159':
            row1 = '부산'
        elif row1 == '108':
            row1 = '서울'
        elif row1 == '133':
            row1 = '대전'

    return row1


def avgtemp (data,place):
    global n
    ycnr = []
    place.append('')
    k = n
    for row in data:
        if row[-1] == ' ':
            continue
        else : 
            try:
                if type(float(row[-3])) == float : 
                    ycnr.append(float(row[-3]))
                    if k == n:
                        row[1] = pcodechange(row[1])
                        place[n] = row[1]
                        n += 1
            except ValueError:
                continue
    return ycnr

koreatemp = avgtemp(koreadata,place)
seoultemp = avgtemp(seouldata,place)
daejeontemp = avgtemp(daejeondata,place)
busantemp = avgtemp(busandata,place)
jejutemp = avgtemp(jejudata,place)

plt.figure(figsize=(20, 10))
plt.plot(koreatemp,label = place[0])
plt.plot(seoultemp,label = place[1])
plt.plot(daejeontemp,label = place[2])
plt.plot(busantemp,label = place[3])
plt.plot(jejutemp,label = place[4])
plt.legend()
plt.title("월별 지역 온도")
plt.xlabel("월")
plt.ylabel("온도ºC")

plt.show()




dif = []
mon = []
for i in range(12):
    
    dif.append(seoultemp[i] - koreatemp[i])
    mon.append(i)
plt.title("전국대비 서울 온도차이")
plt.scatter(mon, dif)
plt.xlabel("월")
plt.ylabel("온도ºC")
plt.show()
#여기도 함수로 만들어보자

dif = []
mon = []
for i in range(12):
    
    dif.append(daejeontemp[i] - koreatemp[i])
    mon.append(i)
plt.title("전국대비 대전 온도차이")
plt.scatter(mon, dif)
plt.xlabel("월")
plt.ylabel("온도ºC")
plt.show()

dif = []
mon = []
for i in range(12):
    
    dif.append(busantemp[i] - koreatemp[i])
    mon.append(i)
plt.title("전국대비 부산 온도차이")
plt.scatter(mon, dif)
plt.xlabel("월")
plt.ylabel("온도ºC")
plt.show()

dif = []
mon = []
for i in range(12):
    
    dif.append(jejutemp[i] - koreatemp[i])
    mon.append(i)
plt.title("전국대비 제주 온도차이")
plt.scatter(mon, dif)
plt.xlabel("월")
plt.ylabel("온도ºC")
plt.show()




koreafile.close()
seoulfile.close()
daejeonfile.close()
busanfile.close()
jejufile.close()


def main():
    print(" ")
if __name__ == '__main__':
    main()


```
